```python
import os
import pandas as pd

# Get the current working directory
current_dir = os.getcwd()

# Specify the folder containing the CSV files
csv_folder = os.path.join(current_dir)

# Get the file paths of all CSV files in the folder
file_paths = [os.path.join(csv_folder, file) for file in os.listdir(csv_folder) if file.endswith('.csv')]

# Create an empty list to store DataFrames
dfs = []

# Load each CSV file into a DataFrame and append to the list
for file_path in file_paths:
    df = pd.read_csv(file_path)
    dfs.append(df)

# Concatenate all DataFrames into one
merged_df = pd.concat(dfs)
```

```python
#Reset the index
merged_df.reset_index(drop=True, inplace=True)
```

```python
# Convert 'started_at' and 'ended_at' columns to datetime
merged_df['started_at'] = pd.to_datetime(merged_df['started_at'])
merged_df['ended_at'] = pd.to_datetime(merged_df['ended_at'])
```

```python
#check for null values
merged_df.isnull().sum()
```

```python
# Check for duplicate rows
duplicate_rows = merged_df[merged_df.duplicated()]

# Print duplicate rows
print("Duplicate Rows:")
print(duplicate_rows)
```

There are some null values but no duplicate rows

### Dealing with Null values

```python
# Extract rows with null values
merged_df[merged_df.isnull().any(axis=1)]
```

```python
# Fill null values in object-type columns with "not specified"
merged_df['start_station_name'].fillna("not specified", inplace=True)
merged_df['start_station_id'].fillna("not specified", inplace=True)
merged_df['end_station_name'].fillna("not specified", inplace=True)
merged_df['end_station_id'].fillna("not specified", inplace=True)

# Fill null values in float-type columns with -999 (or any other placeholder)
merged_df['end_lat'].fillna(-999, inplace=True)
merged_df['end_lng'].fillna(-999, inplace=True)
```

```python
merged_df.isnull().sum()
```

```python
# save the merged DataFrame back to a CSV file
merged_df.to_csv('merged_data.csv', index=False)
```
